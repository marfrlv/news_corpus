```python
# setting up the browser environment
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
driver.get('https://tass.com/')

# wait for 7 seconds
sleep(7)
```

```python
# scrape the articles
headings = []
dates = []
summaries = []
texts = []

for link in tass_links:

    article = NewsPlease.from_url(link)

    heading = article.title
    headings.append(heading)

    date = str(article.date_publish)
    dates.append(date)

    summary = article.description
    summaries.append(summary)

    text = article.maintext
    texts.append(text)

tass_df = pd.DataFrame(
{'Headings': headings,
'Publication dates': dates,
'Summaries': summaries,
'Texts': texts
})

tass_df.to_csv('tass_articles.csv', encoding='utf-8', index = False)
tass_df
```

```python
# extract the publication dates
tass_df = pd.read_csv('tass_articles.csv')
tass_df
```

```python
# find the pattern with re, extract the dates and save them in a list of stings
tass_art_texts = tass_df['Texts']
tass_art_texts_l = list(tass_art_texts)

date_pattern = '[[A-Z]+\,\s[A-z]+\s[0-9]+.'

date_matches = []
for string in tass_art_texts_l:
    match = re.findall(date_pattern, string)
    date_matches.append(match)
print(len(date_matches))
print()
print(date_matches[-1])
dates = [' '.join(inner_list) for inner_list in date_matches]
```

```python
# delete the column with None values and replace it with the extracted dates
tass_df = tass_df.drop('Publication dates', axis = 1)
new_column = pd.Series(dates)
tass_df.insert(loc=1, column='Publication dates', value=new_column)
tass_df
```

```python
# save the new version of the df
tass_df.to_csv('tass_articles_fin.csv', encoding='utf-8', index = False)
```

## The New Yourk Times

```python
# setting up the browser environment
driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
driver.get('https://www.nytimes.com/')

# wait for 7 seconds
sleep(7)
```

```python
# find the show more button and click on it until the end of the page
show_more = driver.find_element(By.XPATH, '//*[@id="site-content"]/div[1]/div[2]/div[2]/div/button')

while True:
    try:
        show_more.click()
    except:
        break
```

```python
# scrape the links
nyt_links = []
elems = driver.find_elements(By.XPATH, '//*[@id="site-content"]/div/div/div/ol/li/div/div/div/a')
links = [elem.get_attribute('href') for elem in elems]
nyt_links.extend(links)
print(len(nyt_links))
```

```python
import json
file = open('nyt_links.json')
nyt_links = json.load(file)
```

```python
headings = []
dates = []
summaries = []
texts = []

from newspaper import Article

for i, link in enumerate(nyt_links):

    try:
        article = Article(link)
        article.download()
        article.parse()

        heading = article.title
        headings.append(heading)

        date = str(article.publish_date)
        dates.append(date)

        summary = article.summary
        summaries.append(summary)

        text = article.text
        texts.append(text)

        print("The number of articles parsed:", i + 1)

    except:
        pass

nyt_df = pd.DataFrame(
{'Headings': headings,
'Publication dates': dates,
'Summaries': summaries,
'Texts': texts
})

nyt_df.to_csv('nyt_articles.csv', encoding='utf-8', index = False)
nyt_df
```
